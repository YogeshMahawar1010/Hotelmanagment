```python
import matplotlib.pyplot as plt
Restaurant = data['Restaurant ID']
Longitude = data['Longitude']
plt.xlabel("Restaurant ID")
plt.ylabel("Longitude")
plt.scatter(Restaurant,Longitude,color = 'yellow')
plt.show()
print("Using Scatter Plot For Location With Longitude")
Restaurant = data['Restaurant ID']
Latitude = data['Latitude']
plt.xlabel("Restaurant ID")
plt.ylabel("Longitude")
plt.scatter(Restaurant, Latitude,color = 'green')
plt.show()
print("Using Scatter Plot For Location With  Latitude")
```

```python
import matplotlib.pyplot as  plt
rating = data['Aggregate rating']
plt.title("Aggregate rating")
plt.plot(rating.describe())
plt.show()
plt.title("Restaurant ID")
plt.hist(data['Restaurant ID'])
plt.show()
           
```
